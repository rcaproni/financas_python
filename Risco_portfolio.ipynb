```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
```

```python
#tickers
tickers = ['ABCB4.SA', 'AZUL4.SA', 'BBAS3.SA', 'BBDC4.SA', 'BRDT3.SA', 'BRSR6.SA', 'CARD3.SA', 'CCRO3.SA', 'COGN3.SA', 'EGIE3.SA', 'EZTC3.SA', 'ITSA4.SA', 'JHSF3.SA', 'LCAM3.SA', 'LEVE3.SA', 'MEAL3.SA', 'MOVI3.SA', 'OIBR3.SA', 'PRIO3.SA', 'RAIL3.SA', 'SAPR11.SA', 'SCAR3.SA', 'SHUL4.SA', 'TAEE11.SA', 'TECN3.SA', 'TRPL4.SA', 'VIVT4.SA', 'VULC3.SA', 'VVAR3.SA']
#weight for each ticker
weights = np.array([0.02265560139, 0.0132791836, 0.06057895305, 0.07859779156, 0.02740391553, 0.02278396174, 0.05603031056, 0.01114239131, 0.02005375752, 0.02545064797, 0.06173348307, 0.06291185143, 0.01239925304, 0.07201550349, 0.03689840447, 0.00921953292, 0.0765272169, 0.003226836522, 0.06523201566, 0.004714186584, 0.03826360845, 0.01533829752, 0.04457771508, 0.03268095205, 0.003309608572, 0.02472036927, 0.02514528278, 0.01688804497, 0.05622132299])
```

```python
#get data in yahoo finance
sec_data = pd.DataFrame()
for t in tickers:
    sec_data[t]=wb.DataReader(t, data_source='yahoo', start='2007-1-1')['Adj Close']
```

```python
sec_data.tail()
```

```python
sec_returns = np.log(sec_data/sec_data.shift(1)) #retorno logaritmico
sec_returns
```

```python
#variancia do portfolio anualizada
port_variance = np.dot(weights.T, np.dot(sec_returns.cov()*250, weights))
port_variance
```

```python
#volatilidade do portfolio
port_vol = (np.dot(weights.T, np.dot(sec_returns.cov()*250, weights)))**0.5
print('Volatilidade do portfolio= '+ str(round(port_vol, 5)*100)+' %')
```
